##📋 Section 3 — Technology Role Filtering

```python
it_roles = [
    "Software",
    "Developer",
    "Programmer",
    "Data Analyst",
    "Data Scientist",
    "Database",
    "Network",
    "Cyber",
    "Security",
    "Cloud",
    "DevOps",
    "QA",
    "Quality Assurance",
    "Machine Learning",
    "Artificial Intelligence",
    "Business Intelligence",
    "UX/UI",
    "Frontend",
    "Backend",
    "Business Analyst",
    "Systems Analyst",
    "IT Project Manager",
    "IT Manager",
    "Product Manager",
    "Agile",
    "Scrum"
]

tech_jobs = df[
    df["Role"].str.contains(
        "|".join(it_roles),
        case=False,
        na=False
    )
].copy()
```

##Section 5 — Create Career Domains

### 🗂️ Career Domain Classification

The original dataset contains hundreds of technology-related job roles.
To enable meaningful career analysis, job roles were grouped into broader technology career domains.

This process is called **career taxonomy creation**, where related occupations are organized into structured categories based on their characteristics and required skills.

The Career Domain feature will support:
- Technology market analysis
- Skill demand comparison
- Career recommendations
- Personalized learning pathways

```python
def classify_domain(role):

    role = role.lower()


    if any(keyword in role for keyword in [
        "backend",
        "frontend",
        "software",
        "developer",
        "programmer",
        "mobile",
        "web"
    ]):
        return "Software Development"


    elif any(keyword in role for keyword in [
        "data analyst",
        "data scientist",
        "business intelligence",
        "machine learning",
        "data architect"
    ]):
        return "Data & Analytics"


    elif any(keyword in role for keyword in [
        "cyber",
        "security",
        "penetration"
    ]):
        return "Cybersecurity"


    elif any(keyword in role for keyword in [
        "cloud",
        "devops"
    ]):
        return "Cloud & DevOps"


    elif any(keyword in role for keyword in [
        "network"
    ]):
        return "Networking"


    elif any(keyword in role for keyword in [
        "database",
        "sql"
    ]):
        return "Database"


    elif any(keyword in role for keyword in [
        "ux",
        "ui",
        "designer",
        "interaction"
    ]):
        return "UI/UX Design"


    elif any(keyword in role for keyword in [
        "business analyst",
        "system analyst",
        "business systems analyst",
        "systems analyst"
    ]):
        return "Business Analysis"

    elif any(keyword in role for keyword in [
    "scrum",
    "agile",
    "product owner",
    "release manager",
    "delivery manager",
    "agile coach"
    ]):
        return "Software Delivery & Agile Management"


    elif any(keyword in role for keyword in [
        "project manager",
        "it manager",
        "product manager",
        "technical product manager"
    ]):
        return "IT Management"


    elif any(keyword in role for keyword in [
        "qa",
        "quality assurance",
        "tester"
    ]):
        return "Quality Assurance"


    else:
        return "Other"
```

```python
tech_jobs["Career Domain"] = tech_jobs["Role"].apply(classify_domain)
```
